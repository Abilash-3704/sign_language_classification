```python
x_tensor = tf.convert_to_tensor(x, dtype=tf.float32)
y_tensor = tf.convert_to_tensor(y, dtype=tf.int32)  # Adjust dtype as needed

# Example: Print shape of tensors
print("Features Tensor Shape:", x_tensor.shape)
print("Labels Tensor Shape:", y_tensor.shape)
```

```python
# Assuming x_tensor and y_tensor are your feature and label tensors respectively
# Convert your dataset into numpy arrays (assuming they are not already)
from sklearn.model_selection import train_test_split
x_numpy = x_tensor.numpy()
y_numpy = y_tensor.numpy()

# Split the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x_numpy, y_numpy, test_size=0.2, random_state=42)

# Print the shapes of the training and testing sets
print(f"Training set: {x_train.shape}, {y_train.shape}")
print(f"Testing set: {x_test.shape}, {y_test.shape}")

x_train = tf.convert_to_tensor(x_train)
x_test = tf.convert_to_tensor(x_test)
y_train = tf.convert_to_tensor(y_train)
y_test = tf.convert_to_tensor(y_test)

# Create TensorFlow datasets from the numpy arrays
train_dataset = tf.data.Dataset.from_tensor_slices((x_train, y_train))
test_dataset = tf.data.Dataset.from_tensor_slices((x_test, y_test))
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np

batch_size = 32

train_dataset = train_dataset.batch(batch_size).shuffle(buffer_size=1000)

# Prepare the testing dataset with batch size
test_dataset = test_dataset.batch(batch_size)

# Reshape the input data and normalize pixel values
train_dataset = train_dataset.map(lambda x, y: (tf.reshape(tf.cast(x, tf.float32) / 255.0, (-1, 28, 28, 1)), y))
test_dataset = test_dataset.map(lambda x, y: (tf.reshape(tf.cast(x, tf.float32) / 255.0, (-1, 28, 28, 1)), y))

sobel_filter_x = np.array([[-1, 0, 1],
                            [-2, 0, 2],
                            [-1, 0, 1]], dtype=np.float32)

sobel_filter_y = np.array([[-1, -2, -1],
                            [0, 0, 0],
                            [1, 2, 1]], dtype=np.float32)

gabor_filter = np.array([[1, 2, 1],
                          [2, 4, 2],
                          [1, 2, 1]], dtype=np.float32)

# Reshape filters to match TensorFlow's expected shape (height, width, in_channels, out_channels)
sobel_filter_x = sobel_filter_x[:, :, np.newaxis, np.newaxis]
sobel_filter_y = sobel_filter_y[:, :, np.newaxis, np.newaxis]
gabor_filter = gabor_filter[:, :, np.newaxis, np.newaxis]

# Define and create your custom CNN model
def create_custom_cnn_model():
    model = models.Sequential()

    # Custom Sobel X filter convolutional layer
    model.add(layers.Conv2D(filters=1, kernel_size=(3, 3), strides=(1, 1), padding='same', 
                            activation='relu', use_bias=False, 
                            kernel_initializer=tf.constant_initializer(sobel_filter_x), 
                            input_shape=(28, 28, 1)))
    
    # Custom Sobel Y filter convolutional layer
    model.add(layers.Conv2D(filters=1, kernel_size=(3, 3), strides=(1, 1), padding='same', 
                            activation='relu', use_bias=False, 
                            kernel_initializer=tf.constant_initializer(sobel_filter_y)))

    # Gabor filter convolutional layer
    model.add(layers.Conv2D(filters=1, kernel_size=(3, 3), strides=(1, 1), padding='same', 
                            activation='relu', use_bias=False, 
                            kernel_initializer=tf.constant_initializer(gabor_filter)))
    
    # Additional convolutional layers for feature extraction
    model.add(layers.Conv2D(32, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))

    model.add(layers.Conv2D(64, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))

    # Flatten the output to feed into dense layers
    model.add(layers.Flatten())

    # Dense (fully connected) layers for classification with dropout
    model.add(layers.Dense(128, activation='relu'))
    model.add(layers.Dropout(0.2))  # Dropout layer with dropout rate of 0.2 (20%)
    model.add(layers.Dense(26, activation='softmax'))  # Output layer with 26 classes (A-Z)

    return model

# Create an instance of the custom CNN model
custom_cnn_model = create_custom_cnn_model()

# Display the model summary
custom_cnn_model.summary()

# Compile the model
custom_cnn_model.compile(optimizer='adam',
                         loss='sparse_categorical_crossentropy',
                         metrics=['accuracy'])

batch_size = 32
epochs = 10

# Train the model using the training dataset
custom_cnn_model.fit(train_dataset, epochs=epochs, verbose=1)

# Evaluate the model on the testing dataset
test_loss, test_accuracy = custom_cnn_model.evaluate(test_dataset, verbose=1)
print(f"Test Accuracy: {test_accuracy}")
```
